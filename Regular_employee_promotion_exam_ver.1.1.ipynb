```python
# 銘柄毎のコメント数の合計値をプロットしてください。
# 流れ
# 1.meigaraとtextが別のテーブルにあるので共通しているmeigara_idをキーにしてテーブルをマージする
# 2.meigara毎のコメント数をカウントする、それを適当な変数に入れる
# 3.その変数を棒グラフでプロットする
```

```python
df1 = pd.merge(df_meigara, df_comment, on="meigara_id")
```

```python
df1.head()
```

```python
df_count = df1.groupby("meigara")["text"].count()
```

```python
print(df_count)
```

```python
df_count.plot.bar(figsize=(150,20))
```

```python
# 見づらいので横棒グラフに変更
df_count.plot.barh(figsize=(5,50))
```

```python
# コメント数が降順になるよう並び替える
# 並び替えるにはsort_values関数を使用
# 降順にするにはascending=True　か　ascending=Falseのどちらか
# 参考:https://news.mynavi.jp/article/zeropython-3/
df_count_sort = df_count.sort_values(ascending=True)
```

```python
# sort_valuesしたものを棒グラフに
df_count_sort.plot.barh(figsize=(5,50))
```

```python
# titleをつけてみる
df_count_sort.plot.barh(title="コメント数", figsize=(5,50))
```

```python
# 凡例をつける
df_count_sort.plot.barh(title="コメント数", label="コメント数", figsize=(5,50))
plt.legend()
```
